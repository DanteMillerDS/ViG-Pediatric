```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from pytorch_grad_cam import GradCAM, HiResCAM, ScoreCAM, GradCAMPlusPlus, AblationCAM, XGradCAM, EigenCAM, FullGrad
from pytorch_grad_cam.utils.model_targets import ClassifierOutputTarget
from pytorch_grad_cam.utils.image import show_cam_on_image
def reshape_transform(tensor, height=7, width=7):
    result = tensor.reshape(tensor.size(0),
        height, width, tensor.size(2))
    result = result.transpose(2, 3).transpose(1, 2)
    return result
def visualize_inference(medical_type, model_type, batch_size, additional_evaluation = False, mean_and_std = False):
  generators, lengths = load_data.create_loader(medical_type, batch_size,finetune = True, mean_and_std = mean_and_std)
  if model_type in ["CycleMLP_B2","resnet50","vig_s_224_gelu","pvig_s_224_gelu","pvt_v2_b2"]:
      classifier = TrainModelClassifier(medical_type,model_type, mean_and_std = mean_and_std)
      classifier.model.load_state_dict(torch.load(f"ViG-Pediatric/results/finetune/{medical_type}/{model_type}/best_model.pth"))
      cams = []
      if additional_evaluation:
        generators, lengths = load_data.create_loader(additional_evaluation, batch_size,finetune = False, mean_and_std = mean_and_std)
        steps = {"Train":lengths["Train"], "Validation":lengths["Validation"], "Test":lengths["Test"]}
        inputs, labels = next(generators[2])
        input = torch.from_numpy(inputs.transpose(0, 3, 1,2)).to(classifier.device)
        label = torch.from_numpy(labels).to(classifier.device).float().unsqueeze(1)
        outputs = classifier.model(input)
        predicted_probs = torch.sigmoid(outputs)
        predicted_labels = (predicted_probs > 0.5).float()
        input = input.detach().cpu().numpy().transpose(0,2,3,1)
        label = label.detach().cpu().numpy()
        predicted_label = predicted_labels.detach().cpu().numpy()
        matches = (predicted_label == label) & (label == 1)
        if matches.any():
          inds = matches.flatten().nonzero()[0]
          print(inds)
          for matching_index in inds:
            input_m = input[matching_index:matching_index+1]
            label_m = label[matching_index]
            predicted_label_m = predicted_label[matching_index]
            print(f"Model {model_type}, Medical Type: {medical_type}, Actual Label: {label_m}, Predicted Label: {predicted_label_m}")
            use_cuda=True
            if model_type == "vig_s_224_gelu" or model_type == "pvig_s_224_gelu":
              grad_cam = GradCam(model=classifier.model, feature_module=classifier.model.prediction, target_layer_names=["4"], use_cuda=use_cuda)
              grayscale_cam = grad_cam(torch.from_numpy(input_m.transpose(0, 3, 1,2)).to(classifier.device), 0)
            elif model_type == "pvt_v2_b2":
              target_layers = [classifier.model.block4[-1].norm1]
              cam = GradCAM(model=classifier.model, target_layers=target_layers,reshape_transform=reshape_transform)
              targets = [ClassifierOutputTarget(0)]
              grayscale_cam = cam(torch.from_numpy(input_m.transpose(0, 3, 1,2)).to(classifier.device), [ClassifierOutputTarget(0)])[0]
            elif model_type == "CycleMLP_B2":
              target_layers = [classifier.model.patch_embed.proj]
              cam = GradCAM(model=classifier.model, target_layers=target_layers)
              targets = [ClassifierOutputTarget(0)]
              grayscale_cam = cam(torch.from_numpy(input_m.transpose(0, 3, 1,2)).to(classifier.device), [ClassifierOutputTarget(0)])[0]
            elif model_type == "resnet50":
              target_layers = [classifier.model.features[7]]
              cam = GradCAM(model=classifier.model, target_layers=target_layers)
              targets = [ClassifierOutputTarget(0)]
              grayscale_cam = cam(torch.from_numpy(input_m.transpose(0, 3, 1,2)).to(classifier.device), [ClassifierOutputTarget(0)])[0]
            image = input_m * mean_and_std["std"] + mean_and_std["mean"]
            image = image.clip(0, 1)
            cam_single = show_cam_on_image(image, grayscale_cam)
            cams.append(cam_single)
            if len(cams) == 2:
              return cams
      return cams
  else:
      print("Did not define a proper classifer!")
mean_and_std = {"ori": {"mean": np.array([0.4856, 0.4856, 0.4856]), "std": np.array([0.2463, 0.2463, 0.2463])}, "ucsd": {"mean": np.array([0.4655, 0.4655, 0.4655]), "std": np.array([0.2523, 0.2523, 0.2523])}}
batch_size = 64
model_types = ["CycleMLP_B2","resnet50","vig_s_224_gelu","pvig_s_224_gelu","pvt_v2_b2"]
model_titles = ["Cycle MLP-B2","ResNet-50","ViG-S","Pyramid ViG-S","PVTv2-B2"]
medical_types = ['ucsd', 'ori']
fig, axs = plt.subplots(2, 5, figsize=(12, 8))
for i in range(len(model_types)):
  cam = visualize_inference(medical_types[0], model_types[i], batch_size, additional_evaluation=medical_types[1], mean_and_std = mean_and_std[medical_types[0]])
  axs[0,i].imshow(cam[0][0])
  axs[1,i].imshow(cam[1][0])
  axs[0,i].axis('off')
  axs[0,i].set_title(f'{model_titles[i]}',fontsize=18)
  axs[1,i].axis('off')
plt.savefig("ucsd_gradcams")
fig, axs = plt.subplots(2, 5, figsize=(12, 8))
for i in range(len(model_types)):
  cam = visualize_inference(medical_types[1], model_types[i], batch_size, additional_evaluation=medical_types[0],mean_and_std = mean_and_std[medical_types[1]])
  axs[0,i].imshow(cam[0][0])
  axs[1,i].imshow(cam[1][0])
  axs[0,i].axis('off')
  axs[0,i].set_title(f'{model_titles[i]}',fontsize=18)
  axs[1,i].axis('off')
plt.savefig("ori_gradcams")
```
